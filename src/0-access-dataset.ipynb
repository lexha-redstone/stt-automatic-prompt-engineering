```python
# English, Mandarin, Cantonese, Korean, Hindi, Japanese, Arabic, Hebrew, Greek
languages = ["en_us", "cmn_hans_cn", "yue_hant_hk", "ko_kr", "hi_in", "ja_jp", "ar_eg", "he_il", "el_gr"]
```

```python
splits = ['dev', 'test', 'train']
base_path = '/Users/lexha/Documents/work/data/fleurs'
df_columns=['id', 'file_name','raw_transcription', 'transcription', 'words', 'num_samples', 'gender']

total_df = pd.DataFrame(columns=['id', 'transcription', 'file_path', 'lang', 'group'])

for lang in languages:
    for g in splits:
        tsv_path = f"{base_path}/{lang}/{g}.tsv"
        df = pd.read_csv(tsv_path, sep='\t', header=None)
        df.columns = df_columns
        df = df.drop_duplicates('id').reset_index(drop=True)
        df = df.drop(['gender', 'num_samples', 'words', 'raw_transcription'], axis=1)
        df['file_path'] = f'/Users/lexha/Documents/work/data/fleurs/{lang}/audio/{g}/' + df['file_name']
        df = df.drop('file_name', axis=1)
        df['lang'] = lang
        df['group'] = g

        print(f"# {lang} / {g} / ROWS : {df.shape}")
        total_df = pd.concat([total_df, df], axis=0, ignore_index=True)
```
